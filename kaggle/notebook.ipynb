```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/audit_benchmarks.csv')
print(df.head(10))
```

```python
plt.figure(figsize=(12, 5))
colors = {'spam': '#ef4444', 'fake': '#f97316', 'offensive': '#8b5cf6', 'irrelevant': '#3b82f6', 'conflict': '#10b981'}
bar_colors = df['violation_type'].map(colors)
plt.bar(df['id'], df['policy_score'], color=bar_colors)
plt.title('BHMarketer — Policy Violation Score by Audit Case', fontsize=14, fontweight='bold')
plt.xlabel('Audit Case ID')
plt.ylabel('Policy Violation Score (0-100)')
plt.tight_layout()
plt.show()
```

```python
violation_avg = df.groupby('violation_type')['recommended_removal'].mean().reset_index()
print('Average Recommended Removals by Violation Type:')
print(violation_avg.to_string(index=False))
```
